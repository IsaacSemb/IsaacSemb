```python
from matplotlib import pyplot as plt
import numpy as np
```

```python

array1 = [ 1,2,3,4,5,6 ]
array2 = [ 11,14,17,22,27,32]

array1 = [ 1,3 ]
array2 = [ 8,10 ]

x = np.array(array1)
y = np.array(array2)

plt.plot(x,y)
plt.show()
```

```python
array1 = [ 1,2,3,4,5,6 ]
array2 = [ 11,14,17,22,27,32]

array1 = [ 1,3 ]
array2 = [ 8,10 ]

x = np.array(array1)
y = np.array(array2)

# to avoid plotting the line you add the o parameter or star
plt.plot(x,y,'*') # this has many parameters
plt.show()

```

```python

# USING MARKERS
  # like stars , O's etc


array1 = [ 1,2,3,4,5,6 ]
array2 = [ 11,14,17,22,27,32]


x = np.array(array1)
y = np.array(array2)

# to avoid plotting the line you add the o parameter or star
plt.plot(x,y,marker = '*') # this has many parameters
plt.show()


```

```python

# changing the line style


# USING MARKERS
  # like stars , O's etc


array1 = [ 1,2,3,4,5,6 ]
array2 = [ 11,14,17,22,27,32]


x = np.array(array1)
y = np.array(array2)


plt.plot(x,y,linestyle='dashed') # this has many parameters
plt.show()

# many like dashed and dotted
```

```python

# setting up labels

array1 = [ 1,2,3,4,5,6 ]
array2 = [ 11,14,17,22,27,32]


x = np.array(array1)
y = np.array(array2)


plt.plot(x,y,linestyle='dashed') # this has many parameters

plt.xlabel('they x axis')
plt.ylabel('the y label')

plt.title('my visualization')


plt.show()



```

```python

# setting up FONTS

array1 = [ 1,2,3,4,5,6 ]
array2 = [ 11,14,17,22,27,32]


# create a dictionary that has 3 keys , family , color , size
font1 = {'family':'serif', 'color':'green', 'size': 40 }
font2 = {'family':'serif', 'color': 'red', 'size': 25 }



x = np.array(array1)
y = np.array(array2)


plt.plot(x,y,linestyle='dashed') # this has many parameters

plt.xlabel('the-x-axis', fontdict=font1)
plt.ylabel('the-y-label', fontdict=font2)

# POSITIONING THE TITLE
# we use the loc function

plt.title('my title-right', loc='right')
plt.title('my title-center', loc='center')



# SHOWING THE GRID LINES ( togle grid on or off )
plt.grid()

plt.show()


```
